```python
def man_name(mfr):
    map={
        "A": 'halal',
        "G": "General mills",
        "K":"Kellogs",
        "N":"Nabisco",
        "Q": "quaker",
        "P":"Post",
        "R":"Ralph"
    }
    return map.get(mfr)

man_name('R')
```

```python
# col is series in pandas
df['mfr'].apply(man_name)
```

```python
df['Manufacturer']= df['mfr'].apply(man_name)
df
```
